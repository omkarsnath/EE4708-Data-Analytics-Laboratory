```python
# importing all necessary libraries

import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
#import previously compiled data file of entire data
numbers = pd.read_csv('train.csv')
numbers.head(10)
```

```python
m = numbers.shape[0] # number of examples
n = numbers.shape[1] # number of pixels
```

```python
# Function to check if the given cell is not nan, and lies within the dataframe limits
def isValid(a, num):
    if((num<784) and (num>=0)):
        if(not (math.isnan(numbers.iloc[a,num]))):
            return True
    return False
```

```python
# Imputation. For every pixel with a missing value, the mean of the 8 pixels adjacent is allotted to it

for i in range(m):
    if(i%100==0):
        print("{}%".format(i/100)) # Progress
    
    for j in range(n):
        if(math.isnan(numbers.iloc[i,j])):
            count = 0 # number of valid pixels
            total = 0 # sum of values of valid neighbour pixels
            for k in [j-29, j-28, j-27, j-1, j+1, j+27, j+28, j+29]:
                # Checks if the pixel is valid & not nan, and then adds the value
                if(isValid(i, k)):
                    count = count + 1
                    total = total + numbers.iloc[i,k]
            
            if(count==0):
                numbers.iloc[i,j] = 0
                # In case to adjacent cells are valid, assigned 0 by default
            else:
                numbers.iloc[i,j] = total/count
```
