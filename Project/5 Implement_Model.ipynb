Program loads the pickle saved model, and makes predictions from the test file as per the given format

```python
# Importing Libraries
import numpy as np
import pandas as pd
import pickle
```

```python
# Loading the test file as per the given format
df=pd.read_csv("public_test.csv",header=None)
```

```python
# Modifying the test file to the required format
test_pixels=[[] for i in range(784)]

for i in range(784):
    for j in range(1000):
        test_pixels[i].append(df[j][i])
test_df=pd.DataFrame()

for i in range(784):
    test_df['pixels'+str(i)]=test_pixels[i]
test_df.head(10)
```

```python
# Loading the model file
model = pickle.load(open("Model.sav", 'rb'))
```

```python
# Making Predictions
test_predict = model.predict(test_df)
```

```python
# Converting Predictions to the required format
solution=pd.DataFrame()
indices=[]
for i in range(1000):
    indices.append('Sample_'+str(i+1)) 
solution['Expected']=test_predict
solution.index=indices
solution.index.name = "Id"
```

```python
# Saving the File
solution.to_csv('Predictions.csv')
```
