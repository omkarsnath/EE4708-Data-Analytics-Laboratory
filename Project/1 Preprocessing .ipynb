```python
# Importing Libraries

import pandas as pd
import glob
import csv
import numpy as np
import math
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Function to extract all the data from the folder of the specific character passed

def extract_csv_files_from_folder(a):
    datapath='.\Training Dataset\character_'+str(a)+'\\'
    allfiles = glob.glob(datapath + '*.csv')
    file_no=0
    
    # Looping through all the files in the folder
    for file in allfiles:   
        print(a,file_no)
        file_no=file_no+1
        Data = pd.read_csv(file,header=None)
        k=0
        for i in range(28):
            for j in range(28):
                pixels[k].append(Data[i][j])
                k=k+1
        pixels[784].append(a) # Adding the class
```

```python
# Declaring the array
pixels=[[] for i in range(785)]
```

```python
#extracting the training data given into one 2-dimensional array
for i in range(10):
    extract_csv_files_from_folder(i)
numbers=pd.DataFrame()
for i in range(784):
    numbers['pixel'+str(i)]=pixels[i]
numbers['label']=pixels[784]
```

```python
# Saving the data to a single file
numbers.to_csv('train.csv', index=False)
```

```python
# Reading the public test file
df=pd.read_csv("public_test.csv",header=None)
df.head(10)
```
